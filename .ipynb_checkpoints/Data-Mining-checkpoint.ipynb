```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import re
```

# Jame Hazfi

```python
df = pd.read_csv('Jam-e-hazfi.csv')
df
```

```python
df_hazfi = pd.DataFrame(columns=['Year', 'Stage', 'team1', 'team2', 'score1', 'score2', 'time', 'date', 'Penalty1', 'Penalty2'])

for i in range(len(df.html)):
    page = df.html[i]
    soup = BeautifulSoup(page, 'html.parser')
    
    match_area = soup.find(class_="match-area")
    
    stage = match_area.select('span.match-stagename')[0].get_text()

    right_name = match_area.select('h2.team-name')[0].get_text()
    left_name = match_area.select('h2.team-name')[1].get_text()

    right_score = match_area.select('span.team-score')[0].get_text()
    left_score = match_area.select('span.team-score')[1].get_text()
    
    penalty_right = 0
    penalty_left = 0

    match_date = match_area.select('span.match-date')[0].get_text()
    match_time = match_area.select('span.match-time')[0].get_text()
    
    if '(' in left_score:
        penalty_right = re.findall(r"\s*\+?(-?\d+)\s*", left_score)[0]
        penalty_left = re.findall(r"\s*\+?(-?\d+)\s*", right_score)[1]

        left_score = re.findall(r"\s*\+?(-?\d+)\s*", left_score)[1]
        right_score = re.findall(r"\s*\+?(-?\d+)\s*", right_score)[0]

    
    new_row = {'Year': df.Year[i], 'Stage': stage, 'team1': left_name, 'team2': right_name, 'score1': left_score,
               'score2': right_score, 'date': match_date, 'time': match_time, 'Penalty1': penalty_left, 'Penalty2': penalty_right}
    df_hazfi = df_hazfi.append(new_row, ignore_index=True)
    
    

df_hazfi
```

```python
df_hazfi.to_csv (r'C:\Users\ssa\Desktop\Data_Tidying\jam-hazfi_result.csv', encoding='utf-8-sig', index = False)
```

# Teams

```python
df = pd.read_csv('teams.csv')
df.html.to_csv (r'C:\Users\ssa\Desktop\Data_Tidying\teams2.csv', encoding='utf-8-sig', index = False, header=True)
df
```

```python
teams_df = pd.DataFrame()
for index, df_row in df.iterrows():
    try:
        row = pd.read_html(df_row['html'], header =0)[0].drop('تیم', axis=1)
    except:
        continue
    row['Competition'] = df_row['Competition']
    row = row[list(row.columns[-1:]) + list(row.columns[:-1])]
    row = row[row['تیم.1'] == df_row.team] 
    
    teams_df = teams_df.append(row, ignore_index=True)
    
teams_df    
```

```python
teams_df.to_csv (r'C:\Users\ssa\Desktop\Data_Tidying\teams_result.csv', encoding='utf-8-sig', index = False)
```

# tables

```python
df = pd.read_csv('tables.csv')
df
```

```python
df_basketball = pd.DataFrame()
df_vollyball = pd.DataFrame()
df_football = pd.DataFrame()

for index, df_row in df.iterrows():
    if 'بسکتبال' in [i.strip() for i in df_row.Competition.split()]:
        row = pd.read_html(df_row['html'], header =0)[0]
        row = row.drop(row.columns[0], axis=1).drop(row.columns[-2:], axis=1)
        row.columns = ['تیم'] + list(row.iloc[0][1:])
        row = row.iloc[1:]
        
        row['Year'] = df_row.Year
        row['Competition'] = df_row.Competition
        row = row[list(row.columns[-2:]) + list(row.columns[:-2])]
        
        df_basketball = df_basketball.append(row, ignore_index=True)
        
    elif 'والیبال' in [i.strip() for i in df_row.Competition.split()]:
        row = pd.read_html(df_row['html'], header =0)[0]
        row = row.drop(row.columns[0], axis=1)
        row.columns = row.iloc[1]
        row = row.iloc[2:]

        row['Year'] = df_row.Year
        row['Competition'] = df_row.Competition
        row = row[list(row.columns[-2:]) + list(row.columns[:-2])]
        
        df_vollyball = df_vollyball.append(row, ignore_index=True)
        
    else:
        row = pd.read_html(df_row['html'], header =0)[0]
        row = row.drop(row.columns[0], axis=1).drop(row.columns[-1], axis=1)
        row.columns = row.iloc[0]
        row = row.iloc[1:]

        row['Year'] = df_row.Year
        row['Competition'] = df_row.Competition
        row = row[list(row.columns[-2:]) + list(row.columns[:-2])]
        
        df_football = df_football.append(row, ignore_index=True)
        
```

```python
df_vollyball.to_csv (r'C:\Users\ssa\Desktop\Data_Tidying\table_vollyball_result.csv', encoding='utf-8-sig', index = False)
df_basketball.to_csv (r'C:\Users\ssa\Desktop\Data_Tidying\table_basketball_result.csv', encoding='utf-8-sig', index = False)
df_football.to_csv (r'C:\Users\ssa\Desktop\Data_Tidying\table_football_result.csv', encoding='utf-8-sig', index = False)
```

# Transfers

```python
df = pd.read_csv('transfers.csv')
df
```

```python
all_transfers = pd.DataFrame()

for index, df_row in df.iterrows():
    soup = BeautifulSoup(df_row.html, 'html.parser')
#     transfers = pd.DataFrame(columns = ['نقل و انتقال' ,'سال' ,'تيم' 
#                                         ,('ورودی', 'قطعی') ,('ورودی', 'احتمالی') ,('ورودی', 'قرضی') ,('ورودی', 'قرضی (بازگشت)') 
#                                         ,('خروجی', 'قطعی') ,('خروجی', 'احتمالی'), ('خروجی', 'قرضی'), ('خروجی', 'قرضی (بازگشت)')])
#     transfers = pd.DataFrame()
    
    stage = re.split(r'[0-9]|-', df_row.transfers)[0]
#     transfers['نقل و انتقال'] = stage
    
    date = re.findall(r'[0-9-]+', df_row.transfers)[0]
#     print(date, "     ", stage)
#     transfers['سال'] = date
    
#     print(date, "     ", stage)
    for row in soup.select('div.m3g-ct-row')[1: ]:
#         print(row)
        transfers = pd.DataFrame()
    
        transfers['نقل و انتقال'] = [stage]

        transfers['سال'] = [date]
    #     print(row)
    
        team_name = row.select('div.ct-team-name')[0].get_text().split('\n')[2]
#         print(team_name.split())
        transfers['تیم'] = team_name
#         print(team_name)

        permanent_in = row.select('div.ct-in span.ct-permanent')
        transfers['ورودی ، قطعی'] = [[i.get_text().split('\n')[1] for i in permanent_in]]
#         print(permanent_in)

        loan_back_in = row.select('div.ct-in span.ct-loan-b')
        transfers['ورودی ، قرضی (بازگشت)'] = [[i.get_text().split('\n')[1] for i in loan_back_in]]
#         print(loan_back_in)

        loan_in = row.select('div.ct-in span.ct-loan')
        transfers['ورودی ، قرضی'] = [[i.get_text().split('\n')[1] for i in loan_in]]
    #     print(load_in)

        likely_in = row.select('div.ct-in span.likely')
        transfers['ورودی ، احتمالی'] = [[i.get_text().split('\n')[1] for i in likely_in]]
    #     print(likely_in)


        permanent_out = row.select('div.ct-out span.ct-permanent')
        transfers['خروجی ، قطعی'] = [[i.get_text().split('\n')[1] for i in permanent_out]]
    #     print(permanent_out)

        loan_back_out = row.select('div.ct-out span.ct-loan-b')
        transfers['خروجی ، قرضی (بازگشت)'] = [[i.get_text().split('\n')[1] for i in loan_back_out]]
    #     print(loan_back_out)

        loan_out = row.select('div.ct-out span.ct-loan')
        transfers['خروجی ، قرضی'] = [[i.get_text().split('\n')[1] for i in loan_out]]
    #     print(loan_out)

        likely_out = row.select('div.ct-out span.likely')
        transfers['خروجی ، احتمالی'] = [[i.get_text().split('\n')[1] for i in likely_out]]
    #     print(likely_out)
    
        all_transfers = all_transfers.append(transfers)
        
all_transfers.reset_index(drop=True)
```

```python
all_transfers.to_csv (r'C:\Users\ssa\Desktop\Data_Tidying\transfers_result.csv', encoding='utf-8-sig', index = False)
```
